```python
#.encode('utf-8')
import HTMLParser
import requests
from bs4 import BeautifulSoup
lists = [0]
same=0
N=0
#Fiest layer
re1=requests.get('http://gametsg.techbang.com/ro/index.php?view=npc&k1=monster&race=Dragon#detail')
re1.encoding = "utf-8"
soup=BeautifulSoup(re1.text, "html.parser")

#Find URL (Second layer)
bid_table = soup.findAll('a',{'class':'color18 mid1'})
for tag in bid_table:
    links = str(['http://gametsg.techbang.com/ro/'+tag['href'].encode('utf-8')])[2:-2]

#Second layer
    re2=requests.get(links)
    re2.encoding = "utf-8"
    soup=BeautifulSoup(re2.text, "html.parser")
    
#Name
    for t in soup.select('.mid1.color8'):
        print "Name:\n"+t.text
        N+=1
    

#Drop&Map
    print "\nDrop&Map:"
    for line in soup.select('.gray_bottom2.h18'):
        for l in line.select('.color8'):
            
            
            for index in range(len(lists)):
                if lists[index]==l.text.encode('utf-8'):
                    same=same+1

            if same==0:
                lists.append(l.text.encode('utf-8'))             #lists 
                print str(lists).decode('string_escape')

            print l.text
            same=0
```

```python
print "\nlists:"
print len(lists)
print N
print type(str(lists).decode('string_escape'))

f = open('Dragon_item.txt','w')
f.write(str(lists).decode('string_escape'))
f.close()
```
